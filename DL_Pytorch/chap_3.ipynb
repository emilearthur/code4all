Scalar -> 0D 
Vector -> 1D 
Matrix -> 2D 
tensor -> 3D. 

A tensor can also be defined as an array: that data structure that sores a collection of numbers that are accessible individually using an index and that can indexed with mutiple indices.

#### Named Tensor 
The dimension (or axes) of tensor are usually index something like pixel location so or color channles. This means when we want to index into a tensor, we need to remember the ordering of the dimensions and write our indexing accordingly. As data is transformed thorugh multiple tensors, keeping track of which dimension contains what data can be error-prone. 

````python
# eg of the statment above 
img_t = torch.randn(3,5,5) # shape [channels, rows, columns]
weights = torch.tensor([0.2126, 0.7152, 0.0722])

print(img_t)
print("\n", weights)
````

````python
# often times we want to generalize, for eg. from grayscal 2D tensor image with height and width dimension to color image with 3D channel as in RGB or from a single image to a batch of images. 

batch_t = torch.randn(2,3,5,5)  # shape [batch, channels, rows, columns]
print(batch_t)
````

````python
# sometimes the RGB channels are in dimension 0, and sometimes they are in dimension 1. But we can generalize by counting from the end: they are always in dimension -3, the third from the end.  

img_gray_naive = img_t.mean(-3) 
batch_gray_naive = batch_t.mean(-3) 

print(img_gray_naive)
print("\n", batch_gray_naive)

print("\n", img_gray_naive.shape, batch_gray_naive.shape)
````

````python
# multiplying things 
unsqueezed_weights = weights.unsqueeze(-1).unsqueeze_(-1) 

img_weights = (img_t * unsqueezed_weights)
batch_weights = (batch_t * unsqueezed_weights)

img_gray_weighted = img_weights.sum(-3) 
batch_gray_weighted = batch_weights.sum(-3) 

print(batch_weights.shape, batch_t.shape, unsqueezed_weights.shape)
````

````python
# using torch einsum for indexing mini-language giving index names to dimensions from sum of such products. 
img_gray_weighted_fancy = torch.einsum('...chw, c->...hw', img_t, weights) 
batch_gray_weighted_fancy = torch.einsum('...chw, c->...hw', batch_t, weights) 
print(batch_gray_weighted_fancy.shape)
````

````python
# named tensors feature 
weights_named = torch.tensor([0.2126, 0.7152, 0.0722], names=['channels'])
print(weights_named)
````

````python
# add names to already existing tensor, we call the refine_names method. 
img_named = img_t.refine_names(..., 'channels','rows','columns') # ellipsis (...) allows you to leave out any number of dimensions
batch_named = batch_t.refine_names(..., 'channels','rows','columns')
print("img named: ", img_named.shape, img_named.names)
print("batch named: ", batch_named.shape, batch_named.names)
````

````python
# the method align_as returns a tensor with missing dimensions added and existing ones permuted to the right other
weights_aligned = weights_named.align_as(img_named)
print(weights_aligned.shape, weights_aligned.names)
````

````python
#using dimension argument to sum. 
gray_named = (img_named * weights_aligned).sum('channels')
print(gray_named.shape, gray_named.names)
````

````python
# combining different elements results to an error 
gray_named = (img_named[..., :3] * weights_named).sum('channel')
````

````python
# if we want use tensor outside function that operate on named tensors, we need to drop the names by renaming them to None.  
gray_plain = gray_named.rename(None) 
print(gray_plain.shape, gray_plain.names)
````

### Tensor elements types

#### Specifying the numeric type with dtype
The ```dtype``` argurment to tensor constructors specifies the numerical data (d) type that will be contained in the tensor. 
The data data type specifies the possibile value the tensor can hold (integers vs point numbers) and the number of bytes per value.  The dtype arg. is deliberately similar to the standard numpy arg. of the same name.   List below 
* torch.float32 or torch.float: 32-bit floating-point 
* torch.float64 or torch.double: 64-bit, double-precision floating-point 
* torch.float16 or torch.half: 16-bit, half-precision floating-point 
* torch.int8: signed 8-bit integers 
* torch.uint8: unsigned 8-bit integers
* torch.int16 or torch.short: signed 16-bit integers
* torch.int32 or torch.int: signed 32-bit integers 
* torch.int64 or torch.long: signed 64-bit integers 
* torch.bool: Boolean 

The default data type for tenosrs is 32-bits floating-point. 

#### dtype for every occasion 
Typical NN are executed with 32-bits floating point precision. Higher precision, like 64 bits does not improve models accuracy and cost more memory and computationtime. 16-bit floating point are not present in standard CPU but are avaiable in GPU, which help decrease the footprint of a neural network if needed with minor impact on accuracy. 

Tensors can be used a indexes in other tensors, in which pytorch expects indexing tensors to have 64-bit interger data type. Creating a tensor with integers as arguments such as using ```torch.tensor([2,2])``` will create a 64-bit tensor by default. 

#### Managing a tensor's dtype attributes
eg. below 


````python
import torch 
double_points = torch.ones(10, 2, dtype=torch.double)
short_points = torch.tensor([[1,2],[3,4]], dtype=torch.short) 
print(short_points.dtype)
````

````python
# we can also cast the output tensor using the corresponding casting method. 
double_points = torch.zeros(10,2).double() 
short_points = torch.ones(10,2).short()

#or a using the to method
double_points = torch.zeros(10,2).to(torch.double) 
short_points = torch.ones(10,2).to(dtype=torch.short)
````

````python
# when mixing input types in operations, the inputs are converted to the larger type automatically. Thus, if we want 32-bit computation, we need to make sure all inputs are 32-bits

points_64 = torch.rand(5, dtype=torch.double) 
points_short = points_64.to(torch.short) 

print(points_64*points_short) # works from Pytorch 1.3 onwards 
````

````python
# Indexing into storage 
points = torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
points.storage()
````

````python
points_storage = points.storage()
print(points_storage[0],points.storage()[1]) # we can index a storage of a 2D indicies, since the layout of a storage is always 1D, regardless of the dimensionaluty of any and all the tensor being refered to. 
````

````python
points = torch.tensor([[4.0,1.0],[5.0,3.0],[2.0,1.0]])
points_storage = points.storage() 
points_storage[0] = 2.0 # we change the storage which leads to change in it contents 

points
````

````python
# modifying stored values: in-place operations 
# Trailing underscore to some tensor operation indicate that the method operation should be in place. For eg. zero_. 
a = torch.ones(3,2) 

print(a) 

a.zero_() 

print(a)
````

````python
# Tensor metadata: Size, offset, and stride. 

# In order to index into a storage, tensor rely on size, offset and stride. 
# The size or shape is a tuple indicating how many elements across each dimension the tensor represents. 
# The storage offset is the index in the storage corresponding to the first element in the tensor.
# The stride is the number of elements in the storage that need to be skipped over to obtain the next element along each dimension.


points = torch.tensor([[4.0, 1.0], [5.0, 3.0], [2.0, 1.0]]) 
second_point = points[1] 
print("Offset: ", second_point.storage_offset()) # index in the second_point.storage correspond first element of second_point 
print("Size: ",second_point.size()) 
print("Shape: ",second_point.shape)
print("Stride", points.stride())
````

````python
print("Size", second_point.size())
print("Offset", second_point.storage_offset())
print("Stride", second_point.stride()) 
````

#### Contiguous Tensors 
A tenosrs whose values are laid out in the storage starting fromt he rightmost dimension onward (ie, moving along rows from a 2D tenosr). Contiguous tensors are convenient because we can visit them efficiently in ord without jumpoing around in the storage. Some tensor operations in PyTorch only works on contiguous tensors, such as view.  

````python
print(points.is_contiguous())
print(points_t.is_contiguous())
````

````python

print("Stride for non-contiguous tensor", points_t.stride(), "\n") 
points_t_cont = points_t.contiguous() # changing a non contiguous tensor to contiguous tensor 
print("Stride for contiguous tensor", points_t_cont.stride()) 

# once we change from non-continguous tensor to contiguous tensor, then the stride wil change. 
````
