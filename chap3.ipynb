```python
# Vectorization for speed 
import math 
import numpy as np 
import time 

n = 10000
a = np.ones(n)
b = np.ones(n) 

class Timer:
    """
    Record multiple running times.
    """
    def __init__(self):
        self.times = [] 
        self.start() 
    
    def start(self):
        # Start the timer
        self.tik = time.time() 
    
    def stop(self):
        # Stop the timer and record the time in a list and return last time
        self.times.append(time.time() - self.tik) 
        return self.times[-1]
    
    def avg(self):
        # Return the average time 
        return sum(self.times)/len(self.times) 
    
    def sum(self):
        # Return the sum of time 
        return sum(self.times)
    
    def cumsum(self):
        # Return the accumulated time.
        return np.array(self.times).cumsum().tolist() 
```

```python

c = np.zeros(n)
timer=Timer()
for i in range(n):
    c[i] = a[i] + b[i]
f'{timer.stop():.5f} sec'
```
